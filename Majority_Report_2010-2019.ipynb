mañana = 5am - 12pm
tarde 12pm - 7pm
noche 7pm - 12am
madrugada 12am - 5am

(Hora >= '05:00:00') and (Hora < '12:00:00') = 'mañana'
(Hora >= '12:00:00') and (Hora < '19:00:00') = 'tarde'
(Hora >= '19:00:00') and (Hora < '24:00:00') = 'noche'
(Hora >= '24:00:00') and (Hora < '05:00:00') = 'madrugada'



```python
data['franja'] = 'madrugada'
data.head()
```

```python
data.head()
```

```python
data.loc[(data['Hora'] >= '06:00:00') & (data['Hora'] < '12:00:00'), 'franja'] = 'mañana'
data.loc[(data['Hora'] >= '12:00:00') & (data['Hora'] < '18:00:00'), 'franja'] = 'tarde'
data.loc[(data['Hora'] >= '18:00:00') & (data['Hora'] < '24:00:00'), 'franja'] = 'noche'
data.loc[(data['Hora'] >= '00:00:00') & (data['Hora'] < '05:00:00'), 'franja'] = 'madrugada'

data.head()
```

```python


# Recorrido por la cantidad de hurtos durante las semanas de los años analizados
timeline = data.groupby(['semana','año']).count()['Cantidad']

timeline.unstack().plot(title='Semanas del Año', legend=True, cmap='Accent')
plt.legend(loc="upper left")
```
