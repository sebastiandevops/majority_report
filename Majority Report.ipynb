```python
df = pd.read_excel('hurto-personas-2019_-_primera_parte_2 (1).xlsx')
```

```python
df1 = pd.read_excel('hurto-personas-2019_-_segunda_parte_1 (1).xlsx')
```

1)Eliminé las primeras filas que no contenían información importante.

```python
df = area_metropolitana2019.drop(range(0, 8, 1),axis=0)
```

```python
print(df.shape)
df.head(10)
```
