```python
# variable para definir los Municipios que hacen parte del área metropolitana del valle de aburrá
area = ['BELLO', 'ENVIGADO', 'ITAGUI', 'MEDELLÍN (CT)', 'LA ESTRELLA', 'SABANETA', 'COPACABANA', 'CALDAS', 'GIRARDOTA', 'BARBOSA']
```

```python
# Esta función la creé para leer los archivos, filtral los datos por los municipios del área metropolitana y reemplazar los valores faltantes por NaN.


def leer_base(archivo, primera=8):
    data = pd.read_excel(archivo, parse_dates=True, header=primera, index_col='Fecha')
    print(data.columns)
    data = data[data['Departamento'] == 'ANTIOQUIA']
    data = data[data['Municipio'].isin(area)]
    data = data.replace('-', np.nan)
    data = data.replace('NO REPORTADO', np.nan)
    data = data.replace('NO REPORTADA', np.nan)
    data = data.replace('NO REPORTA', np.nan)
    data = data.replace('NO DEFINIDO', np.nan)
    return data
```
